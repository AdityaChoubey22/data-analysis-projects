```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("StarWars.csv", encoding="ISO-8859-1")

df = df.drop_duplicates()

df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")


df = df.fillna("Unknown")


print("Dataset Shape:", df.shape)
print("\nColumns:\n", df.columns)



```

```python
df.head()
```

```python
df.columns
```

```python
yes_no = {"Yes": True, "No": False}

for col in [
    "Have you seen any of the 6 films in the Star Wars franchise?",
    "Do you consider yourself to be a fan of the Star Wars film franchise?"
    ]:
    df = df.rename(columns={
    "Have you seen any of the 6 films in the Star Wars franchise?": "seen",
    "Do you consider yourself to be a fan of the Star Wars film franchise?": "fan"
})

df.head()
```

```python


movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in df.columns[3:9]:
    df[col] = df[col].map(movie_mapping)
```

```python
df = df.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

df.head()
```

```python
df[df.columns[9:15]] = df[df.columns[9:15]].apply(pd.to_numeric, errors='coerce')
```

```python
df[df.columns[9:15]].mean()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.bar(range(6), df[df.columns[9:15]].mean())
```

# View counts

It appears that more respondents aw the original movies than the newer movies. This reinforces what we saw in the rankings, where the earlier movies seem to be more popular.

```python
df.columns = df.columns.str.strip()

df = df.rename(columns={
    "What is your gender?": "gender"
})

males = df[df["gender"] == "Male"]
females = df[df["gender"] == "Female"]

print("Male count:", len(males))
print("Female count:", len(females))
```

```python

plt.bar(range(6), males[males.columns[9:15]].mean())
plt.xlabel("Categories")
plt.ylabel("Average Value")
plt.title("Male Data Analysis")
plt.show()


plt.bar(range(6), females[females.columns[9:15]].mean())
plt.xlabel("Categories")
plt.ylabel("Average Value")
plt.title("Female Data Analysis")
plt.show()
```

```python
plt.bar(range(6), males[males.columns[3:9]].sum())
plt.xlabel("Categories")
plt.ylabel("Total Value")
plt.title("Male Data (Sum)")
plt.show()


plt.bar(range(6), females[females.columns[3:9]].sum())
plt.xlabel("Categories")
plt.ylabel("Total Value")
plt.title("Female Data (Sum)")
plt.show()
```

```python

plt.figure()
df['gender'].value_counts().plot(kind='bar')

plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender Distribution")

plt.show()
```

```python

plt.figure()
df['age'].value_counts().sort_index().plot(kind='bar')

plt.xlabel("Age Group")
plt.ylabel("Number of People")
plt.title("Age Distribution")
plt.xticks(rotation=0)

plt.show()
```

```python
plt.figure()
df = df.rename(columns={
    "do_you_consider_yourself_to_be_a_fan_of_the_star_wars_film_franchise?": "fan"
})
df["fan"].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Fan Distribution")
plt.show()
```

```python


cross = pd.crosstab(df['age'], df['fan'])
cross.plot(kind='bar', stacked=True)

plt.title("Age vs Fan Status")
plt.xlabel("Age Group")
plt.ylabel("Count")

plt.show()
```

```python

numeric_df = df.select_dtypes(include=np.number)

if not numeric_df.empty:
    plt.figure()
    sns.heatmap(numeric_df.corr(), annot=True)
    plt.title("Correlation Heatmap")
    plt.show()
```
